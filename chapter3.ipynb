3.1.3 3차 함수 f(x) = (x-2) x (x+2) 그리기

```
# 리스트 2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
  return (x - 2) * x * (x + 2)
```

3.1.6 그래프를 장식하기

```
# 리스트 2-(7)
# 함수를 정의

def f2(x,w):
  return (x - w) * x * (x + 2) # (A) 함수 정의


# x를 정의
x = np.linspace(-3, 3, 100) # (B) x를 100분할하기

#차트 묘사
plt.plot(x, f2(x,2), color='black', label='$w=2$') #(C)
plt.plot(x, f2(x,1), color='cornflowerblue', label='$w=1$') #(D)

plt.legend(loc = "upper left") # (E) 범례 표시
plt.ylim(-15, 15) #(F) y축의 범위
plt.title('$f_2(x)$') # (G) 제목
plt.xlabel('$x$') # (H) x 라벨
plt.ylabel('$y$') # (I) y 라벨
plt.grid(True) # (J) 그리드
plt.show()
```

3.1.7 그래프를 여러 개 보여주기

```
# 리스트 2-(9)
plt.figure(figsize=(10,3)) # (A) figure 지정
plt.subplots_adjust(wspace=0.5, hspace=0.5) # (B) 그래프의 간격을 지정
for i in range(6):
  plt.subplot(2,3, i+1) # (C) 그래프 묘사의 위치를 지정
  plt.title(i + 1)
  plt.plot(x, f2(x,i), 'k')
  plt.ylim(-20, 20)
  plt.grid(True)
plt.show()
```

3.2.1 이변수 함수

```
# 리스트 3-(1)
import numpy as np
import matplotlib.pyplot as plt

# 함수 f3을 정의
def f3(x0, x1):
  r = 2 * x0**2 + x1**2
  ans = r * np.exp(-r)
  return ans

# x0, x1에서 각각 f3을 계산
xn = 9
x0 = np.linspace(-2, 2, xn) #(A)
x1 = np.linspace(-2, 2, xn) #(B)
y = np.zeros( (len(x0), len(x1)) ) #(C)
for i0 in range(xn):
  for i1 in range(xn):
    y[i1, i0] = f3( x0[i0], x1[i1] )
```

```
# 리스트 3-(2)
print(x0)
```

```
# 리스트 3-(3)
print(np.round(y, 1))
```

3.2.3 함수의 표면을 표시 : surface

```
# 리스트 3-(5)
from mpl_toolkits.mplot3d import Axes3D # (A)
xx0, xx1 = np.meshgrid(x0, x1) # (B)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d') #(C)
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black') # (D)
ax.set_zticks( (0, 0.2) ) # (E)
ax.view_init(75, -95) # (F)
plt.show()
```

3.2.4 등고선으로 표시: contour

```
# 리스트 3-(9)
xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros( (len(x0), len(x1)) )
for i0 in range(xn):
  for i1 in range(xn):
    y[i1, i0] = f3( x0[i0], x1[i1] )

xx0, xx1 = np.meshgrid(x0, x1) # (A)

plt.figure(1, figsize=(4,4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') # (B)
cont.clabel(fmt='%3.2f', fontsize=8) # (C)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
